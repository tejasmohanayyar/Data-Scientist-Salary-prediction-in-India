###### Try to find the most common words in job designation based on salary

```python
# let's remove punctuation 
def rem_punct(text): 
    translator = str.maketrans(' ', ' ', string.punctuation) 
    return text.translate(translator) 
```

```python
# Lemmitization
from nltk.stem import wordnet 
from nltk.tokenize import word_tokenize 
lemma = wordnet.WordNetLemmatizer()
nltk.download('wordnet')
# lemmatize string 
def lemmatize_word(text): 
    word_tokens = word_tokenize(text) 
    # provide context i.e. part-of-speech(pos)
    lemmas = [lemma.lemmatize(word, pos ='v') for word in word_tokens] 
    return lemmas 
```

```python
job_desig = {}
for i in train['salary'].unique():
    x = []
    data = train[train['salary']==i]
    data.reset_index(drop = True,inplace = True)
    for j in range(len(data)):
        for k in data['job_desig'][j].split():
            x.append(rem_punct(k.strip().lower()))
    print("{}: {}".format(i,len(set(x))))
    job_desig[i]=list(set(x))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
for i in job_desig.keys():
    plt.figure(figsize = (25,6))
    wordcloud = WordCloud(width = 3024, height = 768, 
                    background_color ='white', 
                    min_font_size = 10).generate(" ".join(job_desig[i])) 
    # Display the generated image:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(i)
    plt.axis("off")
    plt.show()
```
